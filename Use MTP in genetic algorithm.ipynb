```python
def relax_atom(struct, calculator, steps=100, maxstep=.03, fmax=.05):
    struct.set_calculator(calc=calculator)
    dyn = LBFGS(atoms=struct, maxstep=maxstep)
    dyn.run(fmax=fmax, steps=steps)
    
def relax_atoms(atoms, calculator, steps=100, maxstep=.03, fmax=.05):
    """
    Takes in a list of atoms objects and the calculator to use for EFS calculations
    
    returns a list of relaxed structures, achieved with the LBFGS algorithm
    """

    # Loop through atoms. Their energy, forces, and positions are updated in place
    for struct in atoms:
        relax_atom(struct, calculator=calculator, steps=steps, maxstep=maxstep, fmax=fmax)
```

```python
def get_random_composition(prototype, restrictions):
    species = [i for i in restrictions]
    weights = [restrictions[i] for i in restrictions]
    rand_atoms = random.choices(species,
                                weights=weights,
                                k=len(prototype))
    new_struct = prototype.copy()
    new_struct.set_chemical_symbols(rand_atoms)
    return new_struct
```

```python
import operator

        
class Generation:
    def __init__(self, generation, prototype=None, population=None):
        self.generation = generation
        self.prototype = prototype
        self.university_graduation = False
        self.population = population
        
    def set_population(self, population):
        if isinstance(population[0], ase.atoms.Atoms):
            self.population = [Individual(struct, generation=self.generation) for struct in population]
        
        elif isinstance(population[0], Individual):
            self.population = population
    
    def mutation(self):
        """Mutate a portion of the population before they can reproduce""" 
        mutations = np.random.randint(51, size=len(self.population))
        for i, point in enumerate(mutations):

            nl = build_neighbor_list(self.population[i].atoms)
            nn = nl.get_neighbors(point)[0]

            neighbor_dist = 10
            nearest_neighbor = 0
            for neighbor in nn:
                p1 = self.population[i].atoms[point].position
                p2 = self.population[i].atoms[neighbor].position
                dvec = p1 - p2
                dist = np.linalg.norm(dvec)
                if dist < neighbor_dist and dist > 0:
                    neighbor_dist = dist
                    nearest_neighbor = neighbor
                    
            atom_mutate = self.population[i].atoms[point].number
            atom_NN = self.population[i].atoms[nearest_neighbor].number
            
            # Switch the species number
            self.population[i].atoms[point].set_atomic_number(atom_NN)
            self.population[i].atoms[nearest_neighbor].set_atomic_number(atom_mutate)

            
            
#     def fitness(self):
#         """ Calculate the fitness of each individual in the generation """
#         pass

    def university(self, calc, pd):
        """ Sort the population by fitness and luck to determine who gets to reproduce """
        
        
        for struct in self.population:
            struct.relax_struct(calc)
            struct.get_lucky()
            struct.get_vigor(pd)

        vigor = np.array([struct.vigor for struct in self.population])
        luck = np.random.rand(len(vigor)) / 5  #np.array([struct.luck for struct in self.population])
        
        # Normalize the vigor values
        vigor_n = vigor / np.linalg.norm(vigor)
        print(vigor_n)        
        # probability of reproducing
        fitness = vigor_n + luck
        reproduce_zip = zip(fitness, range(len(fitness)))
        
        fittest_individuals = np.flip(sorted(reproduce_zip, key=operator.itemgetter(0)), axis=0)
        self.fittest_individuals = fittest_individuals
    
        # Get fittest individuals
        self.university_graduation = True
        
#         hull_dist = []
#         for struct in initial_population:    
#             pd_entry = PD.PDEntry(struct.get_chemical_formula(), struct.get_potential_energy())
#             e_dist = pd.get_e_above_hull(pd_entry)
#             hull_dist.append(1/e_dist**2)
        
#         hull_norm = np.linalg.norm(hull_dist)
#         hull_normalized = np.array(hull_dist) / hull_norm
#         print(hull_normalized)
    
    def hot_or_not(self):
        """ Allow the fittest individuals to reproduce with each other, leaving the others
        to settle for a partner """
#         self.fittest_individuals
        
        
        couples = []
        for i, struct in enumerate(self.fittest_individuals):
            if i % 2 == 0:
                print(self.fittest_individuals[i])
                p1 = self.population[int(self.fittest_individuals[i][1])]
                p2 = self.population[int(self.fittest_individuals[i+1][1])]

                couples.append([p1, p2])
        
#         top_half = couples[0:len(population)//2]
        print("There are now", len(couples), "couples")
        return couples
    
    def reproduce(self, couple, generation):
        """
        The fittest half of the population will reproduce, and create four offspring
        that are genetically different
           B   b
        A  AB  Ab
        a  aB  ab
        This is achieved by slicing the arrays in a sort of chromosome crossover event
        
        """
#         parent1, parent2 = couple
        children = []
        children_atoms = []

        # Create first pair of children
        crossover = np.random.randint(1, len(parent1)-1)
        
        children_atoms.append(couple[0].atoms[0:crossover+1] + couple[1].atoms[crossover:-1])
        children_atoms.append(couple[1].atoms[0:crossover+1] + couple[0].atoms[crossover:-1])

        # Create second pair of children doing a double crossover
        crossover_A = np.random.randint(1, 25)
        crossover_B = np.random.randint(25, len(parent1)-1)

        children_atoms.append(couple[0].atoms[0:crossover_A+1] 
                              + couple[1].atoms[crossover_A:crossover_B] 
                              + couple[0].atoms[crossover_B:-1])
        
        children_atoms.append(couple[1].atoms[0:crossover_A+1]
                              + couple[0].atoms[crossover_A:crossover_B]
                              + couple[1].atoms[crossover_B:-1])

        for child in children_atoms:
#             child.set_positions(prototype.positions)
#             child.rattle()
            
            new_kid = Individual(atoms=child, generation=generation+1)
            new_kid.set_parents(couple, generation)
            children.append(new_kid)

        return children
    
    
    def get_new_population(self, maxsize=50):
        couples = self.hot_or_not()
        new_population = []
        
        for couple in couples:
            children = self.reproduce(couple, self.generation)
            
            for child in children:
                child.atoms.set_positions(self.prototype.positions)
                child.atoms.rattle()
#                 child.set_generation
                new_population.append(child)
#         print(len(new_population))        
        return new_population[:maxsize]
                
    

    
```

```python
class GenerateVaspInputGR(gvi):
    def __init__(self):
        super().__init__()
    
    def write_kpoints(self, atoms=None, directory='./', autogr=True, **kwargs):
#                       kpdensity=2000, nkpts=None,
#                       kspacing=None, kppra=None, rmin=None):
        """
        The code reads two files, 1) a file the contains the lattice 
        vectors and atomic basis vectors for the crystal (currently 
        only the VASP POSCAR is supported) and 2) a KPGEN file that 
        contains the desired grid density. The code outputs a 
        KPOINTS file containing the irreducible k-points 
        and their weights. The KPGEN file requires one of 
        the following keyword inputs to be present to determine 
        the number of points that will be needed in the grid:
        
        Needs the keyword argument `autogr=True` in order to use the AutoGR
        kpoint routine.
        
        NKPTS (int): the total number of desired k-points.
        KPDENSITY (int): the desired target k-point density.
        KSPACING (list([float float float])): the linear space desired between k-points (same as 
                CASTEP inputs).
        KPPRA (int): the desired number of k-points per reciprocal atom.
        RMIN (float): the minimal distance between points in real space desired.
        (We recommend KPDENSITY or KSPACING.) Only one of these keywords 
                needs to be specified for the code to run.
        """
        print(kwargs)
        
        if autogr is True:
            # Run the `kpoints.x` executable to generate the KPOINTS file
            with open(os.path.join(directory, 'KPGEN'), 'w') as f:
                # Method to determin number of kpoints
                if 'kpdensity' in kwargs:
                    print("KPDENSITY=",str(kwargs['kpdensity']), file=f)
                    
                elif 'nkpts' in kwargs:
                    print("NKPTS=",str(kwargs["nkpts"]), file=f)
                
                elif 'kppra' in kwargs:
                    print("KPPRA=", str(kwargs["kppra"]))

                    
                # Other settings
                if 'rmin' in kwargs:
                    print("RMIN=", str(kwargs['rmin']), file=f)
                if 'kspacing' in kwargs:
                    print("KSPACING=", str(' '.join(kwargs['kspacing'])), file=f)

            sp.run(['.'], executable='kpoints.x', cwd=directory)

        else:
            if atoms is None:
                atoms = self.atoms

            # Don't write anything if KSPACING is being used
            if self.float_params['kspacing'] is not None:
                if self.float_params['kspacing'] > 0:
                    return
                else:
                    raise ValueError("KSPACING value {0} is not allowable. "
                                     "Please use None or a positive number."
                                     "".format(self.float_params['kspacing']))

            p = self.input_params
            with open(join(directory, 'KPOINTS'), 'w') as kpoints:
                kpoints.write('KPOINTS created by Atomic Simulation Environment\n')

                if isinstance(p['kpts'], dict):
                    p['kpts'] = kpts2ndarray(p['kpts'], atoms=atoms)
                    p['reciprocal'] = True

                shape = np.array(p['kpts']).shape

                # Wrap scalar in list if necessary
                if shape == ():
                    p['kpts'] = [p['kpts']]
                    shape = (1, )

                if len(shape) == 1:
                    kpoints.write('0\n')
                    if shape == (1, ):
                        kpoints.write('Auto\n')
                    elif p['gamma']:
                        kpoints.write('Gamma\n')
                    else:
                        kpoints.write('Monkhorst-Pack\n')
                    [kpoints.write('%i ' % kpt) for kpt in p['kpts']]
                    kpoints.write('\n0 0 0\n')
                elif len(shape) == 2:
                    kpoints.write('%i \n' % (len(p['kpts'])))
                    if p['reciprocal']:
                        kpoints.write('Reciprocal\n')
                    else:
                        kpoints.write('Cartesian\n')
                    for n in range(len(p['kpts'])):
                        [kpoints.write('%f ' % kpt) for kpt in p['kpts'][n]]
                        if shape[1] == 4:
                            kpoints.write('\n')
                        elif shape[1] == 3:
                            kpoints.write('1.0 \n')

class VaspGR(Vasp, GenerateVaspInputGR, Calculator):
    pass
    #     def __init__(self):
#         Vasp.__init__(self)
#     pass
# vaspcalc.write_kpoints = write_kpoints
```
